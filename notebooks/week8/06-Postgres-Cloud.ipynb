```python
from langgraph.checkpoint.postgres import PostgresSaver
import psycopg2
from dotenv import load_dotenv
from pathlib import Path
import os
```

### Setup the database

```python
# Connection Test 

# Load environment variables
env_path = Path.cwd().parents[1] / ".env"
print(env_path)
load_dotenv(env_path, override=True)

# Fetch variables
USER = 'postgres.bydnmxkqwjsofgtzmfpl'
PASSWORD = os.getenv("SUPABASE_PASSWORD")
HOST = 'aws-1-us-east-2.pooler.supabase.com'
# PORT = 6543
PORT = 5432
DBNAME = 'postgres'
POOL_MODE = 'session'

# Test Database Connection 
try:
    connection = psycopg2.connect(
        user=USER,
        password=PASSWORD,
        host=HOST,
        port=PORT,
        dbname=DBNAME,
        # pool_mode=POOL_MODE
    )
    print("Connection successful!")
    
    # Create a cursor to execute SQL queries
    cursor = connection.cursor()
    
    # Example query
    cursor.execute("SELECT NOW();")
    result = cursor.fetchone()
    print("Current Time:", result)

    # Close the cursor and connection
    cursor.close()
    connection.close()
    print("Connection closed.")

except Exception as e:
    print(f"Failed to connect: {e}")
```

### Langgraph Checkpointer Table Setup - Run Once

```python

# create langgraph checkpointer setup

conn_string = (
    f"postgresql://{USER}:{PASSWORD}@{HOST}:{PORT}/postgres"
)

with PostgresSaver.from_conn_string(conn_string) as checkpointer:
    checkpointer.setup()

```

### Shopping Cart/Warehouse - Run each sql file once 

```python
# create shopping cart and warehouse tables 

# 1. Configuration - Use the "Session" connection string from Supabase
DB_CONNECTION_STRING = f"postgresql://{USER}:{PASSWORD}@{HOST}:{PORT}/postgres"
# file_path = "../../sql/shopping_cart_table.sql"
file_path = "../../sql/warehouse_management.sql"

try:
    # 2. Connect to the database
    conn = psycopg2.connect(DB_CONNECTION_STRING)
    conn.autocommit = True # Ensures commands run immediately
    
    with conn.cursor() as cursor:
        # 3. Read the SQL file
        with open(file_path, 'r') as f:
            sql_script = f.read()
        
        # 4. Execute the script
        print(f"Executing {file_path}...")
        cursor.execute(sql_script)
        print("Tables created successfully!")

except Exception as e:
    print(f"Error: {e}")
finally:
    if conn:
        conn.close()

```
